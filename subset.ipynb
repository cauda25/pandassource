```python
import pandas as pd
import numpy as np

# NaN 

# df = pd.DataFrame({
#     "a":[4,5,6,6,np.nan],
#     "b":[7,8,np.nan,9,9],
#     "c":[10,11,12,np.nan,12],
# }, index=[1,2,3,4,5])
df = pd.DataFrame({
    "a":[4,5,6,6,np.nan],
    "b":[7,8,np.nan,9,9],
    "c":[10,11,12,np.nan,12],
}, index=list(range(1,6)))

df
```

```python
# 임의로 하나 추출
print(df.sample())
print(df.sample(2))
print(df.sample(frac=0.5))
```

```python
# b 컬럼에서 가장 큰 거 하나의 행

df.nlargest(1,"b")
```

```python
df.nsmallest(2,"a")
```

```python
# pip install seaborn
import seaborn as sns

df = sns.load_dataset("iris")
df.head()

```

```python
# 컬럼 추출
df["petal_length"]
# df.petal_length
```

```python
# 컬럼 두개 이상 추출 시 리스트 형태로

df[["petal_length","sepal_width"]]
```

```python
columns = ["petal_length","sepal_width","sepal_length"]
df[columns].tail(3)
```

```python
# 행,열
# loc[],iloc[] :
df.columns 
# 모든행,컬럼('sepal_length', 'sepal_width', 'petal_length') 추출
# df[['sepal_length', 'sepal_width', 'petal_length']]

# df.loc[행,열] : 행,열 표현시 :
df.loc[:,"sepal_length":"petal_length"]

# 2~5행, 컬럼('sepal_length', 'sepal_width', 'petal_length')
df.loc[2:5,"sepal_length":"petal_length"]

# 2~5행, 컬럼('sepal_length', 'petal_length')
df.loc[2:5,["sepal_length","petal_length"]]
```

```python
#iloc : 인덱스 번호 이용
# 뒤에 인덱스 값은 미포함

# 행 10~19, 열은 전체
df.iloc[10:20]

# 행 1,2,4 열 1,3
df.iloc[[1,2,4],[1,3]]
```

```python
# 행 기준 추출
df[df["petal_length"] > 3]
```

```python
# 행 : sepal_lenth > 5 큰 행만 추출하고, 컬럼은 sepal_length,sepal_width 추출
df.loc[df["sepal_length"] > 5,["sepal_length","sepal_width"]]
```

```python
import pandas as pd

# DataFrame(행,열)

friend_dict_list =[
    {
        "name":"John",
        "age":25,
        "job":"student"
    },
     {
        "name":"Nate",
        "age":35,
        "job":"teacher"
    },
     {
        "name":"Jenny",
        "age":30,
        "job":"developer"
    },
]

friend_df = pd.DataFrame(friend_dict_list)
friend_df
```

```python
# Nate, Jenny  행 만 추출
friend_df.loc[1:2]
friend_df.iloc[1:3]
```

```python
# John, Jenny  행 ,name,job 컬럼추출
friend_df.loc[[0,2],["name","job"]]
friend_df.iloc[[0,2],[0,2]]
```

```python
# 이름이 Nate이고 나이가 25보다 많은 행 추출
friend_df[(friend_df.name == "Nate") & (friend_df.age > 25)]
```

```python
# query()
# pyhton : and,or,not
# pandas : &, | , !


friend_df.query("age>25" and "name=='Nate'")
```
