```python
# cctv

cctv_df = pd.read_excel("./resources/cctv.xlsx",skiprows=2,usecols="C, E:N")
cctv_df
```

```python
# NaN 제거
# cctv_df.dropna()
cctv_df.dropna(inplace=True)
```

```python
# 계 행 제거
cctv_df.drop(index=0,inplace=True)
```

```python
# 컬럼 정리
# 구분 => 구, 별, '2015년 이전 설치된 CCTV ' => 2015년 이전

cctv_df.columns
```

```python
# 컬럼명 변경 : rename()
cctv_df.rename(columns={"구분":"구 별","2015년 이전 설치된 CCTV ":"2015년 이전"},inplace=True)
```

```python
cctv_df.columns
```

```python
# 방법 1) 2020년 이전 컬럼 생성: 2015년 이전 + 2016년 + 2017년 + 2018년 + 2019년 + 2020년

cctv_df["2020년 이전"] = cctv_df["2015년 이전"]+cctv_df["2016년"]+cctv_df["2017년"]+cctv_df["2018년"]+cctv_df["2019년"]+cctv_df["2020년"]
cctv_df.head()
```

```python
cctv_df.drop(columns=["2020년 이전"],inplace=True)
```

```python
# 방법 2) : sum
cctv_df["2020년 이전"] = cctv_df[['2015년 이전', '2016년', '2017년', '2018년', '2019년', '2020년']].sum(axis=1)
cctv_df.head(3)
```

```python
# 필요없는 컬럼 제거: '2015년 이전', '2016년', '2017년', '2018년', '2019년', '2020년'
cctv_df.drop(columns=['2015년 이전', '2016년', '2017년', '2018년', '2019년', '2020년'])
```

```python
cctv_df.drop(columns=cctv_df[cctv_df.columns.difference(['구 별',"2020년 이전",'2021년','2022년', '2023년', '2024년'])],inplace=True)
```

```python
cctv_df.head()
```

```python
# 컬럼 추가
# cctv 총계 = '2021년'+'2022년'+ '2023년'+ '2024년'
cctv_df['cctv 총계'] = cctv_df[["2020년 이전",'2021년','2022년', '2023년', '2024년']].sum(axis=1)
```

```python
cctv_df.head(3)
```

```python
cctv_df = cctv_df[["구 별","2020년 이전",'2021년','2022년', '2023년', '2024년','cctv 총계']]
```

```python
# 구별 cctv 가장 적게 설치되어 있는 구 조회
cctv_df[min(cctv_df['cctv 총계']) == cctv_df['cctv 총계']]

# cctv_df['cctv 총계'].min()
```

```python
# 구별 cctv 가장 많이 설치되어 있는 구 조회
cctv_df[max(cctv_df['cctv 총계']) == cctv_df['cctv 총계']]
```

```python
cctv_df.sort_values("cctv 총계").head(1)
cctv_df.sort_values("cctv 총계",ascending=False).head(1)
```

#### 서울시 거주 인구 데이터 로드

```python
seoul_df = pd.read_excel("./resources/등록인구.xlsx",skiprows=[0,1,3],usecols="B,D,G,j")
seoul_df
```

```python
seoul_df.columns
```

```python
# 컬러명 변경 : 구 별 ,총인구, 한국인 ,외국인
seoul_df.columns=['구 별' ,'총인구', '한국인' ,'외국인']
seoul_df.columns
```

```python
# 어느 구에 외국인이 많이 사는가?
seoul_df[max(seoul_df['외국인']) == seoul_df['외국인']]

seoul_df.sort_values(["외국인"],ascending=False)
```

```python
# 중구 값 변경
# 중 구, 중구
# cctv_df 변경

# 특정 컬럼 조회
# cctv_df['2021년']


# 특정 인덱스 조회
cctv_df.loc[1]
cctv_df.iloc[0]

cctv_df.iloc[1,0]
cctv_df.loc[2,'구 별'] = '중구'
```

```python
# cctv_df 와 seoul_df 병합

merge_df = pd.merge(seoul_df,cctv_df,on='구 별')
```

```python
# 인덱스 변경 : 구 별
# set_index()

merge_df.set_index('구 별',inplace=True)
```

```python
merge_df.head(3)

# cctv 총계 / 총인구 * 100 = 1인당 cctv 비율

merge_df['1인당 cctv 비율'] = merge_df['cctv 총계'] / merge_df['총인구'] * 100
merge_df
```

#### 상관계수
- -1 ~ 1 사이의 값으로 나타남
- 0에 가까울수록 상관 관계는 낮아지며, -1 이나 1에 가까울수록 상관관계는 높아짐
- 증감의 방향이 + 인 경우에는 정정인 상관관계, - 인 경우에는 부정적인 상관관계가 있다고 봄

```python
# cctv 총계와 총인구가 서로 관계가 있는지 파악
merge_df['총인구'].corr(merge_df["cctv 총계"],method='spearman')
```

```python
# 외국인과 cctv 총계 상관관계
merge_df['외국인'].corr(merge_df["cctv 총계"],method='spearman')
```

```python
# 시각화 - 1인당 cctv 비율 
# barh

plt.figure(figsize=(12,8))
plt.barh(y=merge_df.index,width=merge_df['1인당 cctv 비율'])
plt.title("구 별 1인당 CCTV 비율")
```

```python
merge_df['1인당 cctv 비율'].sort_values().plot.barh(figsize=(12,8))
g = plt.title("구 별 1인당 CCTV 비율")
```

```python
plt.figure(figsize=(12,8))
g = sns.barplot(data=merge_df,x="1인당 cctv 비율",y=merge_df.index,order=merge_df['1인당 cctv 비율'].sort_values(ascending=False).index,palette="rocket")
```

#### 5대 범죄

```python
crime_df = pd.read_excel('./resources/5대범죄.xlsx',skiprows=[0,1,2,4],usecols="B,C,D")
crime_df
```

```python
# 컬럼명 : 구 별, 범죄건, 검거건
crime_df.columns = ['구 별','범죄건','검거건']
crime_df.columns 
```

```python
# index 변경 : 구 별
crime_df.set_index('구 별',inplace=True)
```

```python
# 검거율 = 검거건 / 범죄건 * 100
crime_df["검거율"] = crime_df['검거건'] / crime_df['범죄건'] * 100
```

```python
crime_df.head(3)
```

```python
# merge_df 와 crime_df 병합

# merge_df : on => 동일컬럼

merge_crime_df = pd.merge(merge_df,crime_df,left_index=True,right_index=True)
merge_crime_df
```

```python
merge_crime_df.describe()
```

```python
# cctv 총계와 검거간과의 관계
# cctv 총계와 범죄간과의 관계
# scater plot

plt.scatter(x=merge_crime_df["검거건"],y=merge_crime_df['cctv 총계'])
plt.scatter(x=merge_crime_df["범죄건"],y=merge_crime_df['cctv 총계'])
plt.xlabel("범죄건/검거건")
plt.ylabel("cctv 총계")
g = plt.legend(["검거건","범죄건"])
```

```python
ax1 = merge_crime_df.plot.scatter(x="검거건",y='cctv 총계',color='red',s=50)
merge_crime_df.plot.scatter(x="범죄건",y='cctv 총계',ax=ax1,s=50)
plt.legend(["검거건","범죄건"])
g = plt.xlabel("범죄건/검거건")
```

```python
sns.scatterplot(data=merge_crime_df,x='검거건',y='cctv 총계')
sns.scatterplot(data=merge_crime_df,x='범죄건',y='cctv 총계')
```
