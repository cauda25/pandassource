```python
# from chrome import set_chrome_driver
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
# driver.maximize_window()


driver.get("https://www.youtube.com/watch?v=HosW0gulISQ")
time.sleep(1)

# 상품명 입력

time.sleep(3)
# 검색버튼 클릭
# driver.find_element(By.XPATH, '//*[@id="gnb-gnb"]/div[2]/div/div[2]/div[1]/form/div/div/div/div/button[3]').click()
# time.sleep(3)

# 페이지 입력 후 스크롤 이동 : window.scrollTo(시작위치,종료위치)
# driver.execute_script("window.scrollTo(0,2160)")

# 현재 문서 높이
prev_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0,document.documentElement.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.documentElement.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0,0)")

time.sleep(3)

# 댓글 가져오기
response = driver.page_source
soup = BeautifulSoup(response, "html.parser")

```

```python
# 파싱 : 아이디, 코멘트

id_list = soup.select("#author-text > span")
comment_list = soup.select("#content-text > span")
```

```python
for i in range(len(id_list)):
    print(id_list[i].text.strip(), comment_list[i].text.strip())
```

```python
# 엔터 제거 / @ 제거
id_list_clean = []
comment_list_clean = []


for i in range(len(id_list)):
    id,comment = id_list[i].text.strip(), comment_list[i].text.strip()

    cln_id = id.replace("@","")
    cln_id = cln_id.replace("\n","")
    id_list_clean.append(cln_id)

    cln_comment = comment.replace("\t","")
    cln_comment = cln_comment.replace("\n","")
    comment_list_clean.append(cln_comment)

for i in range(len(id_list_clean)):
    print(id_list_clean[i],comment_list_clean[i])
    
```
